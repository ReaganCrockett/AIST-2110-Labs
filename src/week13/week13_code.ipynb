#### My version

````python
# Multiplication Table Ex
from typing import List


def create_mult_table(num: int) -> List[List[int]]:
    table = []
    for i in range(1, num + 1):
        row = []
        for j in range(1, num + 1):
            row.append(i * j)
        table.append(row)
    return table


# Usage
five_table = create_mult_table(5)

for row in five_table:
    print(row)

# Expected output:
# 1	    2	    3	    4	    5
# 2	    4	    6	    8	    10
# 3	    6	    9	    12	    15
# 4	    8	    12	    16	    20
# 5	    10	    15	    20	    25
````

````python
grades = [
    [88, 92, 79],  # Alice
    [95, 85, 91],  # Bob
    [70, 90, 86],  # Cara
]
student_grades = []
student_grades: List[float] = [0] * len(grades)
assign_grades: List[float] = [0] * len(grades[0])
print(student_grades)

for index, student in enumerate(grades):
    for j, assign in enumerate(student):
        student_grades[index] += assign
        assign_grades[j] += assign


print(f'Student Grades: {student_grades}')
print(f'Assignment Grades: {assign_grades}')
# Then, I want sums for each col(assignment) saved to a list then printed out
````
